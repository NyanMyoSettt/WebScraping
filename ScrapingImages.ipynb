```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.geeksforgeeks.org/')
soup = BeautifulSoup(page.content, 'html.parser')
images = soup.find_all('img')

print(soup)
```

```python
images_src = []

for item in images:
    images_src.append(item['src'])

    for index, src in enumerate(images_src):
        if src.endswith('.svg'):
            images_src.pop(index)
images_src.pop(11)
images_src.pop(13)
```

```python
import urllib.request

for index, src in enumerate(images_src):
    images_name_dir =  'images/' + str(index) + '.png'
    try:
        urllib.request.urlretrieve(src, images_name_dir)
    except Exception as error:
        print(error)
        print(index)
```
